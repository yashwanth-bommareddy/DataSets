## Cleaning the data

```python
df = cab_data.merge(transaction_data,on = "Transaction ID")
cab_df = df.merge(customer_data,on = "Customer ID")
cab_df.head()
```

```python
cab_df["Date of Travel"] = pd.to_datetime(cab_df["Date of Travel"],unit = "d", origin = "1899-12-30")
```

```python
cab_df.head()
```

```python
cab_df["Profit on Trip"] = cab_df["Price Charged"] - cab_df["Cost of Trip"]
cab_df.head()
```

```python
cab_df.info()
```

```python
city_data.info()
```

```python
# No null values in both the data frames above. No need to drop any rows or 
#columns as every column will be useful for our analysis
```

```python
cab_df = cab_df.drop_duplicates(subset = ["Transaction ID"])
```

```python
cab_df.info()
```

```python
if city_data['Users'].dtype == 'object':
    city_data['Users'] = city_data['Users'].str.replace(',', '')
city_data['Users'] = city_data['Users'].astype(np.int64)
```

```python
if city_data['Population'].dtype == 'object':
    city_data['Population'] = city_data['Population'].str.replace(',', '')
city_data['Population'] = city_data['Population'].astype(np.int64)
```

```python
city_data.info()
```

```python
# Checking unique values so to elemenate any values entered with a different spelling or in a different format
print(cab_df["Company"].unique())
print(cab_df["City"].unique())
print(cab_df["Payment_Mode"].unique())
print(cab_df["Gender"].unique())
```

```python
# The data is clean with no duplicate values
```

```python
cab_df["Month"] = cab_df["Date of Travel"].dt.strftime("%m-%Y")
cab_df["Year"] = cab_df["Date of Travel"].dt.strftime("%Y")
cab_df.head()
```

## Visualisation

#### Hypothesis 1: There is a correlation between the number of profits earned by companies across different cities.In other words, do some cities tend to generate higher profits for companies compared to others?

```python
df_areax = pd.DataFrame({"City": cab_df["City"], "Year": cab_df["Year"], "Profit": cab_df["Profit on Trip"]})
area_profitx = df_areax.groupby(by=["City","Year"]).mean().unstack()
area_profitx.plot.bar(edgecolor="white", width=0.8,
                      title="Average profit different cities over the years", figsize=(15, 5),
                      xlabel="City", ylabel="Average profit")
```

#### Regarding Hypothesis 1, we found that some cities tend to generate more profits for companies compared to others, indicating a correlation between the number of profits and city of operation. Interestingly, our analysis revealed that the profits earned by companies in these cities remained consistent throughout the years, with very little variation. This suggests that some cities offer more favorable business environments for companies compared to others, possibly due to factors such as a larger customer base, better infrastructure, or lower operating costs.

# 

#### Hypothesis 2: Companies operating in different cities earn similar profits. Specifically, are there any significant differences in profits earned by companies in the same city, or do they earn roughly the same amount of profits regardless of their location?

```python
df_area = pd.DataFrame({"City":cab_df["City"],"Company":cab_df["Company"],"Profit":cab_df["Profit on Trip"]})
area_profit = df_area.groupby(by=["City","Company"]).mean().unstack()
area_profit = area_profit.sort_values(("Profit", "Yellow Cab"), ascending=False)
area_profit.plot.bar(color=["deeppink", "gold"],edgecolor = "white",width = 0.8,
                     title = "Average profit of each company across different cities ",figsize = (15,5),
                     xlabel = "City",ylabel = "Average profit",legend = ["Pink Cab","Yellow Cab"])
```

#### Our analysis found partial support for Hypothesis 2, which suggested that companies operating in different cities earn similar profits. The yellow cab company earned relatively high profits from almost all cities, we observed that  while some cities such as New York, Dallas, and Silicon Valley offered high profits for both the yellow cab and pink cab companies, other cities such as Tucson and Boston offered similar or higher profits to the pink cab company, indicating that profitability can vary depending on the city of operation. These findings suggest that while some cities may offer similar profits to both companies, there are instances where profitability may differ significantly between companies in the same city. This underscores the importance of conducting a detailed analysis of the local market conditions to gain a better understanding of the profitability potential in each city.

# 

#### Hypothesis 3: There is a correlation between the amount of usage by each company in different cities. Is there is a positive relationship between the amount of usage by each company in a given city and their profitability in that same city.

```python
df_area2 = pd.DataFrame({"City":cab_df["City"],"Company":cab_df["Company"],"Usage":cab_df["Transaction ID"]})
area_users = df_area2.groupby(by = ["City","Company"]).count().unstack()
area_users = area_users.sort_values(("Usage", "Yellow Cab"), ascending=False)
area_users.plot.bar(color=["deeppink", "gold"],edgecolor = "white",figsize = (15,5),
                    title = "Amount of usage across different cities",
                    xlabel = "City",ylabel = "Number of times used(2016-2018)",width=0.8)
```

#### Regarding Hypothesis 3, our analysis found that there is no significant correlation between the amount of users of each company in different cities. Although the yellow cab company was found to be the most used company in almost all cities, we observed that the cities with the highest usage of yellow cab were not necessarily the same as those with the highest usage of pink cabs. Additionally, we found that there was no clear relationship between the number of users and profits in most cities. For instance, in cities such as Dallas and Silicon Valley, where profits were relatively high, the usage of cabs was comparatively low.

# 

#### Hypothesis 4: There may be a correlation between the gender of the customer and the choice of cab company used.

```python
df_gender = pd.DataFrame({"Gender":cab_df["Gender"],"Company":cab_df["Company"],"Usage":cab_df["Transaction ID"]})
gender_users = df_gender.groupby(by = ["Gender","Company"]).count().unstack()
gender_users2 = df_gender.groupby(by = ["Company","Gender"]).count().unstack()
fig, axs = plt.subplots(1,2,figsize=(12,6))

gender_users.plot.bar(color=["deeppink", "gold"],edgecolor = "white",
                      title="Amount of usage across different gender",
                      xlabel="Gender", ylabel="Number of times used(2016-2018)",
                      ax=axs[0],zorder = 2)

gender_users2.plot.bar(title="Gender preference towards companies",edgecolor = "white",
                       xlabel="Company", ylabel="Number of times used(2016-2018)",
                       ax=axs[1],zorder = 2)
axs[0].grid(color = "grey", linestyle = "--", linewidth=0.5,axis = "y",zorder = 0)
axs[1].grid(color = "grey", linestyle = "--", linewidth=0.5,axis = "y",zorder = 0)
axs[0].legend(fontsize = 8, loc='upper left')
axs[1].legend(fontsize = 8, loc='upper left')
fig.suptitle('Gender and Company Usage Comparison')
plt.show()
```

#### Our analysis of Hypothesis 4 revealed that there is no significant correlation between the gender of the customer and the cab company preferred. Although yellow cabs were found to be more frequently used by both men and women, we found that the percentage of male and female travelers was almost the same or very close for both the yellow and pink cab companies.

# 

#### Hypothesis 5: There may be a correlation between the age group of the customer and the amount of usage of a specific cab services.

```python
cab_df["Age group"] = pd.cut(cab_df["Age"],[0,24,33,44,60,200],labels = ["<24","24-33","33-44","44-60",">60"])
df_age = pd.DataFrame({"Company":cab_df["Company"],"Transaction ID":cab_df["Transaction ID"],"Age group":cab_df["Age group"]})
age_users = df_age.groupby(by = ["Age group","Company"]).count().unstack()
age_users.plot.bar(color=["deeppink","gold"],edgecolor = "white",
                    title = "Amount of usage by age",
                    xlabel = "Age group",ylabel = "Number of times used(2016-2018)",zorder = 2)
plt.grid(color = "grey", linestyle = "--", linewidth=0.5,axis = "y",zorder = 0)
plt.legend(fontsize = 8, loc='upper right')
```

```python
df_pink = cab_df.loc[cab_df["Company"] == "Pink Cab", ["Company","Age group"]]
df_yellow = cab_df.loc[cab_df["Company"]=="Yellow Cab",["Company","Age group"]]
usage_pink = df_pink.groupby("Age group").count()
usage_yellow = df_yellow.groupby("Age group").count()
fig,axs = plt.subplots(1,2,figsize = (10,5))
axs[0].set_title("Pink Cab usage by different age groups")

axs[1].set_title("Yellow Cab usage by different age groups")
usage_pink.plot.pie(subplots=True,ax = axs[0],autopct = "%1.2f%%",legend = None,ylabel = "")
usage_yellow.plot.pie(subplots=True,ax = axs[1],autopct = "%1.2f%%",legend = None,ylabel = "")
plt.show()
```

#### Our analysis suggests that there is no significant correlation between the age group of the customer and the usage of cab services from a particular company. While the age groups 24-33 and 33-44 account for a relatively higher number of customers for both companies, there is no significant difference in usage across age groups, except for older consumers who use cab services less frequently. Additionally, our analysis indicates that no particular age group is completely inclined towards one specific cab company, as both companies have a similar customer base across all age groups.

# 

#### Hypothesis 6: The overall usage patterns of both cab companies over the years is growing.

```python
pinkcab_df = cab_df.loc[cab_df["Company"] == "Pink Cab",["Month","Company"]]
yellowcab_df = cab_df.loc[cab_df["Company"] == "Yellow Cab",["Month","Company"]]
pinkcab_df["Month"] = pd.to_datetime(pinkcab_df["Month"], format="%m-%Y")
yellowcab_df["Month"] = pd.to_datetime(yellowcab_df["Month"], format="%m-%Y")
pinkcab_growth = pinkcab_df.groupby("Month").count()
yellowcab_growth = yellowcab_df.groupby("Month").count()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
pinkcab_growth.plot.line(title="Usage growth of Pink and Yellow Cabs", ax=ax, color = "deeppink",label="Pink Cab")
yellowcab_growth.plot.line(ax=ax, label="Yellow Cab", color = "gold")
ax.set_ylabel("Amount of usage")
ax.legend(["Pink Cab", "Yellow Cab"])
plt.show()
```

#### We observed a steady increase in the number of users for both cab companies over the years, suggesting that their popularity has been on the rise.

# 

#### Hypothesis 7: The profit growth of each cab company over the years. 

```python
pinkprofit_df = cab_df.loc[cab_df["Company"] == "Pink Cab",["Month","Profit on Trip"]]
yellowprofit_df = cab_df.loc[cab_df["Company"] == "Yellow Cab",["Month","Profit on Trip"]]
pinkprofit_df["Month"] = pd.to_datetime(pinkprofit_df["Month"], format="%m-%Y")
yellowprofit_df["Month"] = pd.to_datetime(yellowprofit_df["Month"], format="%m-%Y")
pinkprofit_growth = pinkprofit_df.groupby("Month").sum()
yellowprofit_growth = yellowprofit_df.groupby("Month").sum()
fig, ax = plt.subplots(figsize=(10, 5))
pinkprofit_growth.plot.line(title="Profit growth of Pink and Yellow Cabs", ax=ax, color = "deeppink",label="Pink Cab")
yellowprofit_growth.plot.line(ax=ax, label="Yellow Cab", color = "gold")
ax.set_ylabel("Profits")
ax.legend(["Pink Cab", "Yellow Cab"])
plt.show()
```

####  It can be observed that both companies have experienced steady profit growth over the years. Although the growth rate is moderate, it is consistent and indicates a positive trend in the financial performance of both companies. However, the profit margin of Yellow cab is relatively very high and it would take considerable timeperiod for Pink cab to reach the same level

```python
pinkretention_df = cab_df.loc[cab_df["Company"] == "Pink Cab",["Year","Customer ID","Transaction ID"]]
yellowretention_df = cab_df.loc[cab_df["Company"] == "Yellow Cab",["Year","Customer ID","Transaction ID"]]
```

```python
pinkretention_df_2016 = pinkretention_df[pinkretention_df["Year"] == "2016"]
pinkretention_df_2017 = pinkretention_df[pinkretention_df["Year"] == "2017"]
pinkretention_df_2018 = pinkretention_df[pinkretention_df["Year"] == "2018"]
yellowretention_df_2016 = yellowretention_df[yellowretention_df["Year"] == "2016"]
yellowretention_df_2017 = yellowretention_df[yellowretention_df["Year"] == "2017"]
yellowretention_df_2018 = yellowretention_df[yellowretention_df["Year"] == "2018"]
pinkr_grouped_2016 = pinkretention_df_2016.groupby("Customer ID").count()
pinkr_grouped_2017 = pinkretention_df_2017.groupby("Customer ID").count()
pinkr_grouped_2018 = pinkretention_df_2018.groupby("Customer ID").count()
yellowr_grouped_2016 = yellowretention_df_2016.groupby("Customer ID").count()
yellowr_grouped_2017 = yellowretention_df_2017.groupby("Customer ID").count()
yellowr_grouped_2018 = yellowretention_df_2018.groupby("Customer ID").count()
pinkr_filtered_2016 = pinkr_grouped_2016[pinkr_grouped_2016["Transaction ID"] > 3]
pinkr_filtered_2017 = pinkr_grouped_2017[pinkr_grouped_2017["Transaction ID"] > 3]
pinkr_filtered_2018 = pinkr_grouped_2018[pinkr_grouped_2018["Transaction ID"] > 3]
yellowr_filtered_2016 = yellowr_grouped_2016[yellowr_grouped_2016["Transaction ID"] > 3]
yellowr_filtered_2017 = yellowr_grouped_2017[yellowr_grouped_2017["Transaction ID"] > 3]
yellowr_filtered_2018 = yellowr_grouped_2018[yellowr_grouped_2018["Transaction ID"] > 3]
```

```python
retention_rate_p2016 = round(pinkr_filtered_2016.shape[0]/pinkr_grouped_2016.shape[0],2)
retention_rate_p2017 = round(pinkr_filtered_2017.shape[0]/pinkr_grouped_2017.shape[0],2)
retention_rate_p2018 = round(pinkr_filtered_2018.shape[0]/pinkr_grouped_2018.shape[0],2)
retention_rate_y2016 = round(yellowr_filtered_2016.shape[0]/yellowr_grouped_2016.shape[0],2)
retention_rate_y2017 = round(yellowr_filtered_2017.shape[0]/yellowr_grouped_2017.shape[0],2)
retention_rate_y2018 = round(yellowr_filtered_2018.shape[0]/yellowr_grouped_2018.shape[0],2)
```

```python
print(retention_rate_p2016)
```

```python
pink_cab_values = [retention_rate_p2016,retention_rate_p2017,retention_rate_p2018]
yellow_cab_values = [retention_rate_y2016,retention_rate_y2017,retention_rate_y2018]
years = [2016, 2017, 2018]
r1 = np.arange(len(years))
r2 = [x + 0.35 for x in r1]
plt.bar(r1, pink_cab_values, color="deeppink", width=0.35, label='Pink Cab',zorder = 2,edgecolor = "white")
plt.bar(r2, yellow_cab_values, color="gold", width=0.35, label='Yellow Cab',zorder = 2,edgecolor = "white")
plt.xticks([r + 0.35/2 for r in range(len(years))], years)
plt.grid(color = "grey", linestyle = "--", linewidth=0.5,axis = "y",zorder = 0)
plt.legend(["Pink Cab","Yellow Cab"],fontsize = 5)
plt.xlabel("Year")
plt.ylabel("Percentage")
plt.title("Rention% of each company over the years")
plt.show()
```

```python
charges_df = pd.DataFrame({"Company":cab_df["Company"],"Price Charged per KM":cab_df["Price Charged"]/cab_df["KM Travelled"],
                            "Year":cab_df["Year"]})
charges_grouped = charges_df.groupby(by = ["Year","Company"]).mean().unstack()
charges_final = charges_grouped.round(2)
charges_final.plot.bar(title = "Price Charged/KM by each company over the years", edgecolor = "white",
                       color = ["deeppink","gold"],zorder = 2)
plt.grid(color = "grey", linestyle = "--", linewidth=0.5,axis = "y",zorder = 0)
plt.legend(fontsize = 5)
plt.show()
```

```python
cost_df = pd.DataFrame({"Company":cab_df["Company"],"Cost per KM":cab_df["Cost of Trip"]/cab_df["KM Travelled"],
                            "Year":cab_df["Year"]})
cost_grouped = cost_df.groupby(by = ["Year","Company"]).mean().unstack()
cost_final = cost_grouped.round(2)
cost_final.plot.bar(title = "Cost incurred/KM by each company over the years", edgecolor = "white",
                       color = ["deeppink","gold"],zorder = 2)
plt.grid(color = "grey", linestyle = "--", linewidth=0.5,axis = "y",zorder = 0)
plt.legend(fontsize = 5, loc='upper left')
plt.show()
```

```python
cab_df["distance group"] = pd.cut(cab_df["KM Travelled"],[0,10,25,40,100],labels = ["<10","10-24","25-39",">39",])
```

```python
distance_df = pd.DataFrame({"Company":cab_df["Company"],"distance travelled":cab_df["distance group"],
                            "Travel distance":cab_df["Transaction ID"]})
distance_grouped = distance_df.groupby(by = ["Company","distance travelled"]).count().unstack()
distance_grouped.plot.bar(title = "Customer usage of each company based on the travel distance", edgecolor = "white",zorder = 2)
plt.grid(color = "grey", linestyle = "--", linewidth=0.5,axis = "y",zorder = 0)
plt.legend(fontsize = 10, loc='upper left')
plt.xticks(rotation=0)
plt.show()
```
